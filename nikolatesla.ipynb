```python
import urllib.request
import re
import matplotlib.pyplot as plt
hand = urllib.request.urlopen('https://en.wikipedia.org/wiki/Nikola_Tesla')
dates = []
for line in hand:
    line = line.decode().strip()
    #d =  re.findall('[^0-9:_=;>a-zA-Z"(-][12]\d{3}', line)
    d =  re.findall('\s[12]\d{3}', line)
    if d != []:
        dates.append(d)
print("found Dates: ")
print(dates)
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

# We can set the number of bins with the *bins* keyword argument.
n_bins = 100
fig, ax = plt.subplots()
ax.hist(dates, bins=n_bins)
ax.set(xlabel='dates', ylabel='count', title='Dates')
plt.show()
    #Your code here
```
